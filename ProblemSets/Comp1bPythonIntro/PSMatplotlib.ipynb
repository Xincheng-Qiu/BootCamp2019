# Introduction to Matplotlib

## Problem 1

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def varmean(n):
    sample = np.random.normal(size=(n,n))
    meanrow = np.mean(sample,1)
    varmean = np.var(meanrow)
    return varmean
```

```python
x = 100*np.arange(1,11)
y = [varmean(n) for n in x]
plt.plot(x,y)
plt.show()
```

## Problem 2

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
sinx = np.sin(x)
cosx = np.cos(x)
arctanx = np.arctan(x)
plt.plot(x, sinx, label = "sin(x)")
plt.plot(x, cosx, label = "cos(x)")
plt.plot(x, arctanx, label = "arctan(x)")
plt.legend()
plt.show()
```

## Problem 3

```python
x1 = np.linspace(-2,0.99,300)
x2 = np.linspace(1.01,6,500)
plt.plot(x1, 1/(x1 - 1), 'm--', linewidth=4)
plt.plot(x2, 1/(x2 - 1),'m--', linewidth=4)
plt.xlim(-2,6)
plt.ylim(-6,6)
plt.show()
```

## Problem 4

```python
x = np.linspace(0, 2*np.pi, 2000)

fig, axes = plt.subplots(2, 2, figsize=[8,8])
plt.suptitle("Problem 4")

ax1 = plt.subplot(221)
ax1.plot(x, np.sin(x),'g-')
plt.title("sin(x)")
plt.axis([0, 2*np.pi, -2, 2])
          
ax2 = plt.subplot(222)
ax2.plot(x, np.sin(2*x),'r--')
plt.title("sin(2x)")
plt.axis([0, 2*np.pi, -2, 2])

ax3 = plt.subplot(223)
ax3.plot(x, 2*np.sin(x),'b--')
plt.title("2sin(x)")
plt.axis([0, 2*np.pi, -2, 2])

ax4 = plt.subplot(224)
ax4.plot(x, 2*np.sin(2*x),'m:')
plt.title("2sin(2x)")
plt.axis([0, 2*np.pi, -2, 2])
```
