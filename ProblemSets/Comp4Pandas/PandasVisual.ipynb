```python
Arbuthnot = data("Arbuthnot")
```

```python
data("Arbuthnot", show_doc=True)
```

```python
Arbuthnot
```

```python
Arbuthnot.index = Arbuthnot["Year"]
```

```python
Arbuthnot.drop(["Year", "Ratio", "Plague", "Mortality", "Total"], axis=1).plot(linewidth=1)
```

```python
Arbuthnot.plot(y=["Females", "Males", "Ratio", "Total"],
               subplots=True, layout=(2,2),
               style=['-', '--', '-.', ':'])
```

```python
Arbuthnot.plot(y=["Mortality", "Plague"])
```

```python
# delete outlier 
rmoutlier = Arbuthnot.loc[Arbuthnot["Mortality"] != max(Arbuthnot["Mortality"])]
```

```python
rmoutlier.plot(kind="scatter", x="Total", y="Mortality", alpha=.8)
```
